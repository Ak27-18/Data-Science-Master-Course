# DATA VISUALIZATION
visualisation on data graphs plots etc

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([1,2,3,4,5])
y=x**2+3
print(x,y)
```

```python
plt.plot(x,y)
```

```python
plt.style.available
```

```python
plt.style.use('seaborn')
plt.title("Market")
plt.plot(x,y,label='Stock Price')
plt.plot(x*3,y+2,label='Predicted')
plt.legend()   #"stock price" wont be displayed unless u use the legend command
plt.xlabel('Period')
plt.ylabel('Price')
plt.show()
```

# Scatter plot

```python
plt.style.use('seaborn-ticks')
plt.title("Market")
plt.scatter(x,y,label='Stock Price')
plt.plot(x*3,y+2,label='Predicted')
plt.legend()   #"stock price" wont be displayed unless u use the legend command
plt.xlabel('Period')
plt.ylabel('Price')
plt.show()
```

# Open CV
installation

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('desktop/ak.jpeg')
#the images shown is in the form of 3D IMAGE i.e a RGB IMAGE.(100,100)3matrix
```

```python
#here img will be a 3D matrix
print(img.shape)
```

```python
plt.imshow(img)
plt.show()
#here the problem is opencv reads the image in BGR format
```

```python
#to dispaly in RBG format by switching
rgb_image= cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
#gray_image=cv2.cvtColor(img,cv2,COLOR_BGR2GRAY)
```

```python
plt.imshow(rgb_image)
plt.axis("off")
plt.show()
```

```python
#to CROP images
crop_image=rgb_image[:,105:250,:]   #[rows,columns,channels]
plt.imshow(crop_image)
plt.axis("off")
plt.show()
```

```python
crop_image.shape
```

```python
total_pixels=crop_image.shape[0]*crop_image.shape[1]   #225*120
print(total_pixels)
#and each image will have 2700 pixels and 3 rgb soeach pixel will have 51000
```

```python
#resizing the image
resized_image=cv2.resize(crop_image,(100,100))
plt.imshow(resized_image)
plt.axis("off")
plt.title("AK"+str(resized_image.size)+str(resized_image.shape))
plt.show()
#reduced to lower resolution and lower size
```

# Pandas-Working with Tabular Data
how to work with tables, using pandas to manage data like excell shits and "csb" files reading the data

```python
#5 Attribute ranging from 1 to 100 
pokemons={
    "height":np.random.randint(1,100,5),
    "weight":np.random.randint(50,150,5),
    "power":np.random.randint(200,300,5)
}
```

```python
pokemons
```

The data here is random based on the attributes and the range specified

```python
#storing the data in sheet by using pandas
import pandas as pd
df=pd.DataFrame(pokemons)
print(df)
```

much cleaner way to display data.

```python
#.head()comment will give the first 5 values, unless specified otherwise
df.head()
```

```python
#to know what attributes we are dealing with we need to print columns
print(df.columns)
```

# Creat a CSV(comma Separated) Files
